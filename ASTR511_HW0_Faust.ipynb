# Recreate color-color plots which show how to discriminate between QSO, Galaxies, and stars

## Load and clean data

```python
from astropy.io import fits 
from astropy.table import Table
import numpy as np

# QSO data file
QSOdata = fits.getdata('dr7qso.fit', 1)
QSO_data_tabular = Table(QSOdata)
#QSO_data_tabular[0:3] # Get var names

# Extract and clean relevant magnitudes
Raw_QSO_data = QSO_data_tabular['GTMAG', 'RTMAG', 'ITMAG', 'JMAG', 'HMAG', 'KMAG']
nz_index = (Raw_QSO_data['GTMAG'] != 0)*(Raw_QSO_data['RTMAG'] != 0)*(Raw_QSO_data['ITMAG'] != 0)*(Raw_QSO_data['JMAG'] != 0)*(Raw_QSO_data['HMAG'] != 0)*(Raw_QSO_data['KMAG'] != 0)
QSO_Mags = Raw_QSO_data[nz_index]
del Raw_QSO_data
QSO_Mags[0:5]
```

```python
# Star data file
N = 1000 #read a bunch of lines
from itertools import islice
import numpy as np
with open("SDSS2MASS_fixedRv_b30.dat", "r") as starfile:
    head = list(islice(starfile, N+1))
    
starfile.close()
    
Star_Mags = np.ones((N, 6))

for x in range(N):  
        Star_Mags[x, 0] = float(head[x+1][153:162]) # best fit g
        Star_Mags[x, 1] = float(head[x+1][163:172]) # best fit r 
        Star_Mags[x, 2] = float(head[x+1][173:182]) # best fit i 
        Star_Mags[x, 3] = float(head[x+1][440:449]) # best fit J 
        Star_Mags[x, 4] = float(head[x+1][450:459]) # best fit H 
        Star_Mags[x, 5] = float(head[x+1][460:469]) # best fit K   
    

#  each of the fits differ by enough to shift plot. Should have asked which to use.  
#        Star_Mags[x, 0] = float(head[x+1][510:519]) # best fit g
#        Star_Mags[x, 1] = float(head[x+1][520:529]) # best fit r 
#        Star_Mags[x, 2] = float(head[x+1][530:539]) # best fit i 
#        Star_Mags[x, 3] = float(head[x+1][550:559]) # best fit J 
#        Star_Mags[x, 4] = float(head[x+1][560:569]) # best fit H 
#        Star_Mags[x, 5] = float(head[x+1][570:579]) # best fit K

Star_Mags[0:2, 0:6]
```

```python
# Galaxies file

SDSS_Gxy_data = fits.getdata('object_sdss_imaging.fits', 1)
twoMASS_Gxy_data = fits.getdata('object_twomass.fits', 1) 
Gxy_SDSS_tabular = Table(SDSS_Gxy_data)
Gxy_2MASS_tabular = Table(twoMASS_Gxy_data)

#Gxy_SDSS_tabular['PSFFLUX']
Gxy_2MASS_tabular[0:3]
```

```python
# Extract relevant magnitudes and fluxes  SDSS flux always nonzero, apparently
Raw_Gxy2MASS_data = Gxy_2MASS_tabular['J_M_K20FE', 'H_M_K20FE', 'K_M_K20FE']
nz_index = (Raw_Gxy2MASS_data['J_M_K20FE'] != 0)*(Raw_Gxy2MASS_data['H_M_K20FE'] != 0)*(Raw_Gxy2MASS_data['K_M_K20FE'] != 0)
SDSS_PSF_flux = Gxy_SDSS_tabular[nz_index]['PSFFLUX']
Gxy_SDSSmags = 22.5 - 2.5 * np.log10(SDSS_PSF_flux)
# Gxy_SDSSmags = -2.5 * np.log10(SDSS_PSF_flux / 3631.)

#Gxy_Len = sum(nz_index)
#Gxy_Len
Gxy_2MASSmags = Raw_Gxy2MASS_data[nz_index]
Gxy_2MASSmags[0:3]
#Gxy_Len = sum(nz_index)
len(Gxy_SDSSmags), len(Gxy_2MASSmags)
#Gxy_SDSSmags
```

## Create color-color plot

```python
# Plot a color-color diagram
from matplotlib import pyplot as plt
%matplotlib inline
N = 500 #number of points to plot 

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
#top-left plaquette
ax[0][0].scatter(QSO_Mags[0:N]['RTMAG']-QSO_Mags[0:N]['ITMAG'], QSO_Mags[0:N]['GTMAG']-QSO_Mags[0:N]['RTMAG'], s=1, color='blue', label="QSOs")
ax[0][0].scatter(Star_Mags[0:N, 0]-Star_Mags[0:N, 1], Star_Mags[0:N, 1]-Star_Mags[0:N, 2], s=1, color='red', label="Stars")
ax[0][0].scatter(Gxy_SDSSmags[0:N, 0]-Gxy_SDSSmags[0:N, 1], Gxy_SDSSmags[0:N, 1]-Gxy_SDSSmags[0:N, 2], s=2, color='black', label="Galaxies")
ax[0][0].legend(loc=2)
ax[0][0].set_xlim(-0.5, 2.)
ax[0][0].set_ylim(-0.5, 2.)
ax[0][0].set_xlabel('$g^{*}-r^{*}$', fontsize=20)
ax[0][0].set_ylabel('$r^{*}-i^{*}$', fontsize=20)

#bottom-left plaquette
ax[1][0].scatter(QSO_Mags[0:N]['RTMAG']-QSO_Mags[0:N]['ITMAG'], QSO_Mags[0:N]['GTMAG']-QSO_Mags[0:N]['RTMAG'], s=1, color='blue')
ax[1][0].scatter(Star_Mags[0:N, 3]-Star_Mags[0:N, 4], Star_Mags[0:N, 4]-Star_Mags[0:N, 5], s=1, color='red')
ax[1][0].scatter(Gxy_2MASSmags[0:N]['J_M_K20FE']-Gxy_2MASSmags[0:N]['H_M_K20FE'], Gxy_2MASSmags[0:N]['H_M_K20FE']-Gxy_2MASSmags[0:N]['K_M_K20FE'], s=2, color='black')
ax[1][0].set_xlim(-0.25, 1.25)
ax[1][0].set_ylim(-0.25, 1.25)
ax[1][0].set_xlabel('$J-H$', fontsize=20)
ax[1][0].set_ylabel('$H-K_{s}$', fontsize=20)

#top-right plaquette
ax[0][1].scatter(QSO_Mags[0:N]['GTMAG']-QSO_Mags[0:N]['ITMAG'], QSO_Mags[0:N]['JMAG']-QSO_Mags[0:N]['KMAG'], s=1, color='blue')
ax[0][1].scatter(Star_Mags[0:N, 0]-Star_Mags[0:N, 2], Star_Mags[0:N, 3]-Star_Mags[0:N, 5], s=1, color='red')
ax[0][1].scatter(Gxy_SDSSmags[0:N, 0]-Gxy_SDSSmags[0:N, 2], Gxy_2MASSmags[0:N]['J_M_K20FE']-Gxy_2MASSmags[0:N]['K_M_K20FE'], s=2, color='black')
ax[0][1].set_xlim(-1., 4.)
ax[0][1].set_ylim(0., 2.5)
ax[0][1].set_xlabel('$g^{*}-i^{*}$', fontsize=20)
ax[0][1].set_ylabel('$J-K_{s}$', fontsize=20)

#bottom-right plaquette
ax[1][1].scatter(Star_Mags[0:N, 0]-Star_Mags[0:N, 2], Star_Mags[0:N, 2]-Star_Mags[0:N, 5], s=1, color='red')
ax[1][1].scatter(QSO_Mags[0:N]['GTMAG']-QSO_Mags[0:N]['ITMAG'], QSO_Mags[0:N]['ITMAG']-QSO_Mags[0:N]['KMAG'], s=1, color='blue')
ax[1][1].scatter(Gxy_SDSSmags[0:N, 0]-Gxy_SDSSmags[0:N, 2], Gxy_SDSSmags[0:N, 2]-Gxy_2MASSmags[0:N]['K_M_K20FE'], s=2, color='black')
ax[1][1].set_xlim(-1, 4)
ax[1][1].set_ylim(0., 5.)
ax[1][1].set_xlabel('$g^{*}-i^{*}$', fontsize=20)
ax[1][1].set_ylabel('$i^{*}-K_{s}$', fontsize=20)

fig.savefig('fits_level2.png', bbox_inches='tight')
```
